```python
import matplotlib
import pip
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv('/content/store.csv')

# Display the first few rows of the DataFrame to understand its structure
print(df.head())

# Extract data for the graph (adjust column names)
x_data = df['CompetitionDistance']
y_data = df['Promo2SinceWeek']

# Create a scatter plot
plt.scatter(x_data, y_data, label='Scattered Plot')

# Add labels and title
plt.xlabel('X-axis Assortment ')
plt.ylabel('Y-axis promo2')
plt.title('sccarated Plot from CSV Data')

# Add legend
plt.legend()

# Show the plot (you can also save it using plt.savefig)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv('/content/store.csv')

# Display basic information about the dataset
print("Dataset Overview:")
print(df.info())

# Display summary statistics for numerical columns
print("\nSummary Statistics:")
print(df.describe())

# Display the first few rows of the DataFrame
print("\nFirst Few Rows:")
print(df.head())

# Explore data visually (customize based on your data types)
# Example: Histogram for numerical columns
df.hist(figsize=(10, 8))
plt.suptitle('Histograms of Numerical Columns', y=0.95)
plt.show()

# Example: Count plot for categorical columns
categorical_columns = df.select_dtypes(include='object').columns
for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    df[col].value_counts().plot(kind='bar', color='skyblue')

    plt.title(f'Count Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```
